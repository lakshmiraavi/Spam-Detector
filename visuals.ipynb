```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
ham_words = pd.read_csv('dataset/ham_words_count.csv', index_col=0).reset_index()
# ham_words = ham_words[ham_words.ham_words != 'xkeywords']
ham_words.to_csv('dataset/ham_words_count.csv', index=False)
ham_words.head(50)
```

```python
ham_words[['ham_words', 'count']].tail(10)
# ham_words.plot(kind='hist', figsize=(12,8))
# ham_words.plot(kind='bar', figsize=(12,8))
```

```python
# ham_words.plot(kind='hist', figsize=(12,8))
plt.figure(figsize=(15,10))
plt.scatter(ham_words.index.tolist(),ham_words['count'].tolist(), color="orange", marker=",")
```
